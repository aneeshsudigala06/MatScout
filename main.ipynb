```python
# Import Pandas and Matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import Data
data = pd.read_csv('ncaa_data.csv')
```

```python
# Find Count of Wrestlers by Weight Class
weight_class_counts = data["weight class"].value_counts()
```

```python
# Plot a Bar Graph of Wrestlers by Weight Class
weight_class_counts.sort_index().plot(kind='bar', color='#CD6155')
plt.title('Wrestlers by Weight Class')
plt.xlabel('Weight Class')
plt.ylabel('Number of Wrestlers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count Wrestlers by State
state_counts = data['state'].value_counts()
```

```python
# Plot a Bar Graph of Wrestlers by State
state_counts.plot(kind='barh', color='#AF7AC5')
plt.title('Wrestlers by State')
plt.xlabel('Number of Wrestlers')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
# Group Data by Weight Class and Placement
placement_data = data.groupby(['weight class', 'placement']).size().unstack(fill_value=0)
```

```python
# Plot a Bar Graph of Wrestlers by Weight Class and Placement
placement_data.plot(kind='bar', stacked=True, colormap='tab20b_r')
plt.title('Wrestlers by Placement and Weight Class')
plt.xlabel('Weight Class')
plt.ylabel('Number of Wrestlers')
plt.legend(title='Placement', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count Wrestlers by High School
high_school_counts = data['high school'].value_counts().head(10)
```

```python
# Plot a Bar Graph of Top 10 High Schools Producing Wrestlers
high_school_counts.plot(kind='barh', color='#5499C7')
plt.title('Top 10 High Schools Producing Wrestlers')
plt.xlabel('Number of Wrestlers')
plt.ylabel('High School')
plt.tight_layout()
plt.show()
```

```python
# Filter Top Placements
top_placements = data[data['placement'] == 1]['state'].value_counts()
```

```python
# Plot a Pie Chart of Top Placements
top_placements.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.tab20b_r.colors)
plt.title('Placement Success Rate by State')
plt.ylabel('')  # Hide y-axis label
plt.tight_layout()
plt.show()
```
